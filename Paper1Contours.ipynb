```python
%matplotlib widget
import pickle
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Circle as circ
import matplotlib.patches as patches

plt.rc('mathtext', fontset='stix')
plt.rc('font', family='STIXGeneral')
plt.rc('figure', autolayout=True)
```

## E3 Contour Plot

```python
plt.close('all')
fig = plt.gcf()
ax = fig.gca()

# Colors chosen by eye to look good together
colors = np.array(['#FF7F7F', '#dedad2', '#FFC6B2', '#98B3BD', '#6A85A9', '#E0CFAF', '#808982'])

"""
Draws the circles manually based on running the main code at the chosen L1, L2, L3 value. Since we didn't have the analytical expression 
at this time this was just typed in manually and fit to the data by eye.

Since only L3 below 0.5 for E3 has edge circles, I just divided the two functions and manually drew the edge circles.

They are both defined as circs(radius, color of fill, L3 value for legend)
The comment on the side of each circle is the fraction of points in the circle (from running the base code, and the L3 value associated with that result.
"""
circs(0, colors[0], 0.4)        #0.198 frac, L3 = 0.4
circedge(0.3, colors[0],  0.4)

circs(0, colors[1], 0.45)        #0.292 frac, L3 = 0.45
circedge(0.22, colors[1],  0.45)

circs(0.65, colors[2],  0.5)

circs(0.575, colors[3],  0.6)    #0.498 frac, L3 = 0.6


circs(0.43, colors[4],  0.8)   # 0.719 frac, L3 = 0.8

circs(0.32, colors[5],  0.9)  # 0.850 frac, L3 = 0.9

circs(0.15, colors[6],  0.98) # 0.968 frac, L3 = 0.98



plt.rcParams['legend.title_fontsize'] = 15
plt.rcParams['font.size'] =  18
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles[::-1], labels[::-1], title = r'$L_B  /  L_{\rm LSS}$', loc = (1.01,0.09), ncol = 1, fontsize = 18, title_fontsize=20)

plt.text(2.1, 1.8, r'$E_3$', fontsize = 35)
plt.text(2.1, 0.05, r'$L_A = 2L_{\rm LSS}$')
plt.xlim([0,2])
plt.ylim([0,2])
plt.yticks([0.5, 1.0, 1.5, 2.0])
ax.set_aspect(1)

plt.show()
```

```python
def circs(k, color, L3):
    """
    Draws the circles in the 'core' points of E3 (the corners and center of Fundamental Domain
    The locations can be found analytically or from just looking at the results from the base code
    """
    
    midCirc = circ((1,1),k, color = color, fill = True, label = f'{L3}')
    topLeft = circ((0,2),k, color = color, fill = True)
    topRight = circ((2,2),k, color = color, fill = True)
    botLeft = circ((2,0),k, color = color, fill = True)
    botRight = circ((0,0),k, color = color, fill = True)
    return(ax.add_patch(midCirc), ax.add_patch(topLeft), ax.add_patch(topRight), ax.add_patch(botLeft), ax.add_patch(botRight))
```

```python
def circedge(k, color, frac):
    """
    Same as above for the edge circles
    """
    
    left = circ((0,1),k, color = color, fill = True)
    right = circ((2,1),k, color = color, fill = True)
    bot = circ((1,0),k, color = color, fill = True)
    top = circ((1,2),k, color = color, fill = True)
    return(ax.add_patch(left), ax.add_patch(right), ax.add_patch(bot), ax.add_patch(top))
```

```python
def circsE4(k, color, opac, L3):
    midCirc = circ((0,0),k, color = color, fill = True, alpha = opac, label = f'{L3}')
    botRight = circ((2,0),k, color = color, fill = True, alpha = opac)
    topLeft = circ((0,1.15),k, color = color, fill = True, alpha = opac)
    topRight = circ((1,0.58),k, color = color, fill = True, alpha = opac)
    botLeft = circ((-1,1.75),k, color = color, fill = True, alpha = opac)
    other = circ((1,1.75),k, color = color, fill = True, alpha = opac)
    return(ax.add_patch(midCirc), ax.add_patch(botRight), ax.add_patch(topLeft), ax.add_patch(topRight), ax.add_patch(botLeft), ax.add_patch(other))
```

```python
def circsE5(k, color, L3):
    midCirc = circ((0,0),k, color = color, fill = True, label = f'{L3}')
    botRight = circ((2,0),k, color = color, fill = True)
    botLeft = circ((-1,1.75),k, color = color, fill = True)
    other = circ((1,1.75),k, color = color, fill = True)
    return(ax.add_patch(midCirc), ax.add_patch(botRight), ax.add_patch(botLeft), ax.add_patch(other))
```

```python
def circedgeE5(k, color, L3):
    bot = circ((1,0.58),k, color = color, fill = True, label = f'{L3}')
    top = circ((0,1.15),k, color = color, fill = True)
    return( ax.add_patch(bot), ax.add_patch(top))
```
